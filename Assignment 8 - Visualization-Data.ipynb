# In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
ds = sns.load_dataset('iris')
ds.head()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
features
```

```python
x = ds.loc[:,features].values
x
```

```python
y = ds.loc[:,['species']].values
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pc = pca.fit_transform(x)
pdf = pd.DataFrame(data = pc,columns = ['Principal Component 1', 'Principal Component 2', 'Principal Component 3'])
pdf
```

```python
finalDF = pd.concat([pdf, ds[['species']]], axis=1)
finalDF
```

```python
x_vals = finalDF['Principal Component 1']
y_vals = finalDF['Principal Component 2']
z_vals = finalDF['Principal Component 3']
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

targets = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDF['species'] == target
    ax.scatter(finalDF.loc[indicesToKeep, 'Principal Component 1']
               , finalDF.loc[indicesToKeep, 'Principal Component 2']
               , finalDF.loc[indicesToKeep, 'Principal Component 3']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.show()
```
